Agregando datasets

```python
file_paths = [
    "Documents/coleta2013.csv",
    "coleta2023_amostrado.csv",
]
```

```python
coletatotal = pd.DataFrame()

for file_path in file_paths:
    data_df = pd.read_csv(file_path, sep=',',error_bad_lines=False)
    coletatotal = coletatotal.append(data_df)
```
